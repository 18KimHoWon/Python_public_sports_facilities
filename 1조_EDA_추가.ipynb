```python
import pandas as pd
```

```python
df = pd.read_csv('서울시 공공체육시설 정보_0423정리_추가.csv')
```

```python
df
```

# 데이터 준비

```python
# 시설 대관여부 : 가능 > 1, 불가능 > 0으로 변경
df['대관여부'] = 0

for i in range(len(df)) :
    if df['시설대관여부'][i] == '가능' :
        df['대관여부'][i] = 1
```

```python
# 시설 주소, 시설대관여부, 시설편의시설 열 제거
df.drop(columns=['시설주소', '시설대관여부'], inplace=True)
```

```python
df = df.astype({'샤워실여부':'int', '탈의실여부':'int','라커룸여부':'int', '매점여부':'int', '휴게시설여부':'int'})
```

```python
df.head(3)
```

```python
df.info()
```

# EDA : 구별 공공체육시설 비교

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

## 개수

```python
# 구별로 시설의 개수를 세는 데이터프레임 생성
gu_count = df.groupby('시군구').count()['시설명'].reset_index()
gu_count
```

```python
# 컬럼명 : 시설명 > 개수
gu_count = gu_count.rename(columns={'시설명':'개수'})
```

```python
plt.figure(figsize=(4,4))
ax = sns.barplot(data=gu_count, x='시군구', y='개수')
ax.bar_label(ax.containers[0])
plt.title('구별 공공체육시설 개수')
plt.show()
```

=> 노원구가 압도적으로 공공체육시설의 개수가 많았고, 강북구가 공공체육시설의 개수가 가장 적었다.

### 구별 데이터 나누기 + 시설 개수 열추가

#### 시설 개수 열 추가

```python
# 원래 데이터프레임 df와 구별 시설의 개수를 나타낸 데이터프레임 gu_count를 합쳐서 df1를 만든다
df1 = pd.merge(df, gu_count, how='left', on='시군구')
df1.head(3)
```

#### 강북구

```python
gangbuk = df1[df1['시군구']=='강북구'].reset_index()
gangbuk.drop(columns=['index'], inplace=True)
```

```python
gangbuk.head(3)
```

#### 노원구

```python
nowon = df1[df1['시군구']=='노원구'].reset_index()
nowon.drop(columns=['index'], inplace=True)
```

```python
nowon.head(3)
```

#### 도봉구

```python
dobong = df1[df1['시군구']=='도봉구'].reset_index()
dobong.drop(columns=['index'], inplace=True)
```

```python
dobong.head(3)
```

## 시설종류

```python
plt.figure(figsize=(8,4))
ax = sns.countplot(data=df1, x='시설종류')
ax.bar_label(ax.containers[0])
plt.xticks(rotation=45)
plt.title('공공체육시설 종류별 개수')
plt.show()
```

=> 강북구, 노원구, 도봉구 전체의 공공체육시설 종류를 살펴봤을 때는 생활체육관, 수영장, 배드민턴장이 많았다.

### 강북구

```python
plt.figure(figsize=(8,4))
ax = sns.countplot(data=gangbuk, x='시설종류')
ax.bar_label(ax.containers[0])
plt.xticks(rotation=45)
plt.title('강북구 공공체육시설 종류별 개수')
plt.show()
```

=> 강북구 내에서는 생활체육관, 수영장, 배드민턴장의 개수가 많은 편으로 전체와 비슷한 경향을 보였다. 하지만 전체를 살펴봤을때보다 시설의 종류가 적었다.

### 노원구

```python
plt.figure(figsize=(8,4))
ax = sns.countplot(data=nowon, x='시설종류')
ax.bar_label(ax.containers[0])
plt.xticks(rotation=45)
plt.title('노원구 공공체육시설 종류별 개수')
plt.show()
```

=> 노원구 내에서는 생활체육관, 수영장, 축구장, 배드민턴장의 개수가 많은 편으로 전체와 비슷한 경향을 보였다. 또한 시설의 종류도 많고, 각 개수도 충분했다.

### 도봉구

```python
plt.figure(figsize=(8,4))
ax = sns.countplot(data=dobong, x='시설종류')
ax.bar_label(ax.containers[0])
plt.xticks(rotation=45)
plt.title('도봉구 공공체육시설 종류별 개수')
plt.show()
```

=> 도봉구 내에서는 생활체육관, 수영장, 배드민턴장의 개수가 많은 편으로 전체와 비슷한 경향을 보였다. 하지만 전체를 살펴봤을때보다 시설의 종류가 적은 편이었다.

## 운영시간

```python
opening_hours = df1.groupby('시군구').mean()[['평일_운영시간', '토요일_운영시간', '일요일_운영시간']].reset_index()
```

```python
opening_hours
```

```python
plt.subplot(131)
ax1 = sns.barplot(data=opening_hours, x='시군구', y='평일_운영시간')
ax1.bar_label(ax1.containers[0])
plt.title('평일 평균 운영시간')

plt.subplot(132)
ax2 = sns.barplot(data=opening_hours, x='시군구', y='토요일_운영시간')
ax2.bar_label(ax2.containers[0])
plt.title('토요일 평균 운영시간')

plt.subplot(133)
ax3 = sns.barplot(data=opening_hours, x='시군구', y='일요일_운영시간')
ax3.bar_label(ax3.containers[0])
plt.title('일요일 평균 운영시간')

plt.tight_layout()
plt.show()
```
